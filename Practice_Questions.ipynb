```python
df["Profit"] = (df["Unit_Price"] - df["Unit_Cost"]) * df["Quantity_Sold"]
print(df)
```

```python
df.nlargest(10, "Profit")
```

```python
# 1. Convert Sale_Date to datetime format if not already done
df["Sale_Date"] = pd.to_datetime(df["Sale_Date"])

# 2. Group by month and sum the Sales_Amount
monthly_trends = df.groupby(df["Sale_Date"].dt.to_period("M"))["Sales_Amount"].sum()

print(monthly_trends)
```

```python
monthly_trends.plot(kind="line", marker="o", title="Monthly Sales Trends")
```

```python
df["Sale_Date"] = pd.to_datetime(df["Sale_Date"])

# --- 1 & 2. KPI Metrics ---
total_sales = df["Sales_Amount"].sum()
total_profit = df["Profit"].sum()
print(f"=== FINANCIAL SUMMARY ===")
print(f"Total Sales: ${total_sales:,.2f}")
print(f"Total Profit: ${total_profit:,.2f}\n")

# --- 3. Region-wise Sales ---
plt.figure(figsize=(8, 4))
region_sales = df.groupby("Region")["Sales_Amount"].sum().reset_index()
sns.barplot(data=region_sales, x="Region", y="Sales_Amount",hue="Region", palette="Blues_d",legend=False)
plt.title("Region-wise Sales", fontsize=14, weight="bold")
plt.ylabel("Total Sales ($)")
plt.show()

# --- 4. Category-wise Sales ---
plt.figure(figsize=(8, 4))
cat_sales = df.groupby("Product_Category")["Sales_Amount"].sum().reset_index()
sns.barplot(data=cat_sales, x="Product_Category", y="Sales_Amount",hue="Product_Category", palette="Greens_d",legend =False)
plt.title("Category-wise Sales", fontsize=14, weight="bold")
plt.ylabel("Total Sales ($)")
plt.show()

# --- 5. Monthly Sales Trend ---
plt.figure(figsize=(10, 4))
monthly_sales = df.groupby(df["Sale_Date"].dt.to_period("M"))["Sales_Amount"].sum()
monthly_sales.index = monthly_sales.index.to_timestamp()
plt.plot(monthly_sales.index, monthly_sales.values, marker="o", color="orange", linewidth=2)
plt.title("Monthly Sales Trend", fontsize=14, weight="bold")
plt.ylabel("Sales ($)")
plt.xticks(rotation=45)
plt.show()

# --- 6. Sales by Sales Representative ---
plt.figure(figsize=(8, 4))
rep_sales = df.groupby("Sales_Rep")["Sales_Amount"].sum().sort_values(ascending=False).reset_index()
sns.barplot(data=rep_sales, x="Sales_Amount", y="Sales_Rep",hue="Sales_Amount", palette="Purples_r",legend=False)
plt.title("Sales by Sales Representative", fontsize=14, weight="bold")
plt.xlabel("Total Sales ($)")
plt.show()

# --- 7. Customer Type Distribution ---
plt.figure(figsize=(5, 5))
cust_dist = df["Customer_Type"].value_counts()
plt.pie(cust_dist, labels=cust_dist.index, autopct="%1.1f%%", colors=["#ff9999","#66b3ff"], startangle=90)
plt.title("Customer Type Distribution", fontsize=14, weight="bold")
plt.show()

# --- 8. Payment Method Distribution ---
plt.figure(figsize=(8, 4))
sns.countplot(data=df, x="Payment_Method",hue="Payment_Method", palette="Pastel1",legend=False)
plt.title("Payment Method Distribution", fontsize=14, weight="bold")
plt.xticks(rotation=15)
plt.show()

# --- 9. Sales Channel Distribution ---
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x="Sales_Channel",hue="Sales_Channel", palette="Pastel2",legend=False)
plt.title("Sales Channel Distribution", fontsize=14, weight="bold")
plt.show()

# --- 10. Discount Analysis ---
plt.figure(figsize=(8, 4))
sns.histplot(data=df, x="Discount", kde=True, color="purple", bins=20)
plt.title("Discount Analysis (Distribution of Discounts Given)", fontsize=14, weight="bold")
plt.xlabel("Discount Percentage")
plt.show()
```
